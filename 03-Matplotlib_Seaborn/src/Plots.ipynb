```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df = pd.read_csv("C:/Users/PC/Desktop/Ciencias de datos/Blog/03-Matplotlib_Seaborn/data/cars_sales.csv")
```

```python
warnings.filterwarnings("ignore")
```

```python
_ = plt.hist(df.price)
plt.show()
```

```python
sns.set_style(style = "whitegrid")

mean,median = df['price'].mean(),df['price'].median()

fig,(ax) = plt.subplots(1,1,figsize = (20,8))


ax.set_title("Custom Histogram")
ax.hist(df['price'],ec="k",color="#FADA5E",lw=3)


ax.axvline(mean,
           color="red",
           linestyle="--",
           lw=3,label="Mean")


ax.axvline(median,
           color="blue",
           linestyle="--",
           lw=3,label="Median")

ax.legend()

plt.show()
```

```python
serie_mean_price = df.groupby('manufacturer')['price'].mean().sort_values(ascending = True)
```

To facilitate the interpretation of the bar graph, we calculated the average value according to the car manufacturer.

```python
serie_mean_price = df.groupby('manufacturer')['price'].mean().sort_values(ascending = True)
serie_mean_price.plot(kind = 'bar')
plt.show()
```

```python
serie_mean_price.plot(kind = 'barh',title = 'Mean Price')
plt.show()
```

```python
fig = plt.subplots(1,1,figsize = (20,8))
colors = ['#fdfd96','#fdcae1','#84b6f4','#b0c2f2', '#ff6961','#77dd77','#87CEEB','#ff8000','#9ee000']
_= serie_mean_price.plot(kind = 'barh',title = 'Mean Price',
                         color = colors)

plt.show()
```

The list of parameters is similar, with the difference of parameter kind, in this case we will use "barh", it allows to create a horizontal bar graph, it makes the graph more comfortable to visualize.
